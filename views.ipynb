```python
import os, json
auto_dir = 'auto-docs'  # where we keep any auto-generated content
doc_dir = os.path.join(auto_dir, 'api')  # the main page for all documentation
ref_dir = os.path.join(doc_dir, 'references')
```

```python
def get_api(language, subsection_url=None):
    ''' Return all the info needed to dynamically generate an API page
    '''
    ## load language-specefic reference
    ref_file = os.path.join(ref_dir, "{}.json".format(language))
    with open(ref_file) as f:
        reference = json.load(f)
    if subsection_url:
        subsection = None
        for sub in reference['subsections']:
            for sub_sub in sub['subsections']:
                if sub_sub['relative_url'] == subsection_url:
                    subsection = sub_sub
        if subsection:
            examples = subsection.pop('subsections')
            return subsection, examples
    else:
        sections = {subsection['id']:subsection['name'] for subsection in reference['subsections']}
        return sections  
```
