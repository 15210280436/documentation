```python
import os, json
publish_dir = 'published'
doc_dir = os.path.join(publish_dir, 'api-docs')  # the main page for all documentation
ref_dir = os.path.join(doc_dir, 'references')
```

```python
def get_api(language, branch_url=None):
    ''' Return all the info needed to dynamically generate an API page
    '''
    ## load language-specefic reference
    ref_file = os.path.join(ref_dir, "{}.json".format(language))
    with open(ref_file) as f:
        reference = json.load(f)
    if branch_url:
        return_branch = None
        for branch in reference['branches']:
            for sub_branch in branch['branches']:
                if sub_branch['relative_url'] == branch_url:
                    return_branch = sub_branch
        if return_branch:
            examples = return_branch.pop('branches')
            return return_branch, examples
    else:
        return reference['branches']

def get_urls():
    urls = set()
    for ref_file in os.listdir(ref_dir):
        language, ext = os.path.splitext(ref_file)
        with open(os.path.join(ref_dir, ref_file)) as f:
            reference = json.load(f)
        for branch in reference['branches']: # 'chart-types', 'layout', etc...
            for sub_branch in branch['branches']: # e.g.: area, histogram, etc.
                if 'relative_url' in sub_branch:
                    path = [language, sub_branch['relative_url']]
                    urls.add('/'.join(part.strip('/') for part in path))
    url_list = list(urls)
    url_list.sort()
    return url_list
```
