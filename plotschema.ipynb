```python
def get_trace_schema(trace):
    trace_schema = {}
    trace_schema ['name'] = i
    trace_schema ['permalink'] = 'reference/#'+i
    if p['schema']['traces'][i]['meta']:
        trace_schema ['description'] = p['schema']['traces'][i]['meta']['description']
    else: pass
    return trace_schema

def get_attribute_schema(trace, path=None, link_path=None):
    att_schema=[]
    nested_atts=[]
    if not path:
        path = p['schema']['traces'][trace]['attributes']
    for att in path:
        if 'src' not in att:
            try:
                attribute = {}
                attribute ['name'] = trace+' > '+att
                if link_path:
                    attribute ['permalink'] = 'reference/#'+trace+'-'+link_path+'-'+att
                else:
                    attribute ['permalink'] = 'reference/#'+trace+'-'+att
                attribute ['description'] = path[att]['description']
                att_schema.append(attribute)
            except:
                attribute = {}
                attribute ['name'] = trace+' > '+att
                if link_path:
                    attribute ['permalink'] = 'reference/#'+trace+'-'+link_path+'-'+att
                else:
                    attribute ['permalink'] = 'reference/#'+trace+'-'+att
                attribute ['description'] = 'Properties for '+att
                att_schema.append(attribute)
    return att_schema

def get_nested_schema(trace, path=None, link_path=None):
    nested_atts=[]
    if not path:
        path = p['schema']['traces'][trace]['attributes']
    for att in path:
        if 'src' not in att:
            if isinstance(path[att], dict):
                for i in path[att]:
                    if isinstance(path[att][i], dict):
                        if att not in nested_atts:
                            nested_atts.append(att)
    return nested_atts
```

```python
schema = []
def get_all(trace):
    schema.append(get_trace_schema(trace))
    schema.append(get_attribute_schema(trace))
    n_atts = get_nested_schema(trace)
    for n_att in n_atts:
        schema.append(get_attribute_schema(trace,
                                           path=p['schema']['traces'][trace]['attributes'][n_att],
                                           link_path=n_att))
    return schema
        
        
        
    
    
```
