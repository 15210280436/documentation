```python
import json
p = json.load(open('_data/plotschema.json'))

schema = []
```

### 1st Level: Traces 

```python
for i in p['schema']['traces']:
    print i
```

Add trace dictionaries to schema array.
The trace dictionary include name: trace name, permalink: reference/#trace-name, and description if applicable. 

```python
for i in p['schema']['traces']:
    trace = {}
    trace ['name'] = i
    trace ['permalink'] = 'reference/#'+i
    if p['schema']['traces'][i]['meta']:
        trace ['description'] = p['schema']['traces'][i]['meta']['description']
    else: pass
    schema.append(trace)
schema
```

### Level 2: Attributes

```python
for i in p['schema']['traces']:
    second_attributes = []
    for att in p['schema']['traces'][i]['attributes']:
        if 'src' not in att:
            try:
                attribute = {}
                attribute ['name'] = i+' > '+att
                attribute ['permalink'] = 'reference/#'+i+'-'+att
                attribute ['description'] = p['schema']['traces'][i]['attributes'][j]['description']
                schema.append(attribute)
            except:
                attribute = {}
                attribute ['name'] = i+' > '+att
                attribute ['permalink'] = 'reference/#'+i+'-'+att
                attribute ['description'] = 'Properties for '+att
                schema.append(attribute)            
schema
```
